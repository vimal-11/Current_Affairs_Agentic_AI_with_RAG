```python
xml_bytes = dicttoxml.dicttoxml(top_headlines, custom_root='article')

# 3. Decode the byte string to a regular string and parse it for pretty-printing.
xml_string = xml_bytes.decode('utf-8')
dom = parseString(xml_string)
pretty_xml = dom.toprettyxml()

# 4. Print the formatted XML to the console (optional).
print("--- Generated XML ---")
print(pretty_xml)

# 5. Save the pretty XML to a file.
file_name = "article.xml"
with open(file_name, "w", encoding="utf-8") as f:
    f.write(pretty_xml)

print(f"\nSuccessfully saved XML to {file_name}")
```
